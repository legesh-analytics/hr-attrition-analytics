```python
import pandas as pd
import numpy as np
```

```python
n = 2000
# ------------------------------------------------------------------------
employee_id = np.arange(1, n+1)
employee_name = ["Emp_" + str(i).zfill(4) for i in employee_id]
# ------------------------------------------------------------------------
gender = np.random.choice(["Male", "Female"], size=n)
# ------------------------------------------------------------------------
age = np.random.randint(22, 55, size=n)
def age_group(a):
    if a < 25:
        return "Early Career"
    elif a < 35:
        return "Young Professional"
    elif a < 45:
        return "Mid Career"
    elif a < 55:
        return "Senior"
    else:
        return "Pre-Retirement"

age_group_list = [age_group(a) for a in age]
# ------------------------------------------------------------------------
departments = ["Sales", "HR", "Technology", "Finance", "Operations", "Support"]
department = np.random.choice(departments, size=n)
# ------------------------------------------------------------------------
locations = ["Mumbai", "Bangalore", "Hyderabad", "Pune", "Chennai"]
location = np.random.choice(locations, size=n)
# ------------------------------------------------------------------------
bands = ["B2", "B3", "B4", "B5", "B6"]
band = np.random.choice(bands, size=n)
# ------------------------------------------------------------------------
def assign_job(b):
    if b == "B2":
        return "Analyst"
    elif b == "B3":
        return np.random.choice(["Analyst", "Senior Analyst"])
    elif b == "B4":
        return np.random.choice(["Engineer", "Senior Analyst"])
    elif b == "B5":
        return np.random.choice(["Senior Engineer", "Manager"])
    else:
        return "Manager"

job_title = [assign_job(b) for b in band]
# ------------------------------------------------------------------------
manager = ["Mgr_" + dept + "_" + str(np.random.randint(1,4)) for dept in department]
# ------------------------------------------------------------------------
def salary(b):
    if b == "B2":
        return np.random.randint(30000, 60000)
    elif b == "B3":
        return np.random.randint(60000, 100000)
    elif b == "B4":
        return np.random.randint(100000, 150000)
    elif b == "B5":
        return np.random.randint(150000, 250000)
    else:
        return np.random.randint(250000, 400000)

base_salary = [salary(b) for b in band]
# ------------------------------------------------------------------------
sources = ["Referral", "Job Portal", "Agency", "Campus", "Internal"]
source_of_hire = np.random.choice(sources, size=n)
# ------------------------------------------------------------------------
df = pd.DataFrame({
    "EmployeeID": employee_id,
    "EmployeeName": employee_name,
    "Gender": gender,
    "Age": age,
    "AgeGroup": age_group_list,
    "Department": department,
    "Location": location,
    "Band": band,
    "JobTitle": job_title,
    "ManagerName": manager,
    "BaseSalary": base_salary,
    "SourceOfHire": source_of_hire
})
# ------------------------------------------------------------------------
```

```python
dept_ot_mean = {
    "Sales": 18,
    "HR": 8,
    "Technology": 14,
    "Finance": 10,
    "Operations": 24,
    "Support": 22
}
```

```python
def generate_performance(b):
    if b == "B2":
        return np.random.choice([2, 3, 4], p=[0.15, 0.65, 0.20])
    elif b == "B3":
        return np.random.choice([2, 3, 4, 5], p=[0.10, 0.60, 0.25, 0.05])
    elif b == "B4":
        return np.random.choice([2, 3, 4, 5], p=[0.08, 0.55, 0.28, 0.09])
    elif b == "B5":
        return np.random.choice([2, 3, 4, 5], p=[0.07, 0.50, 0.30, 0.13])
    else:
        return np.random.choice([2, 3, 4, 5], p=[0.05, 0.50, 0.30, 0.15])

df["BasePerformanceRating"] = df["Band"].apply(generate_performance)
```

```python
np.random.seed(42)

df["BaseOvertimeHours"] = df["Department"].map(dept_ot_mean) + np.random.normal(0, 4, size=len(df))
df["BaseOvertimeHours"] = df["BaseOvertimeHours"].clip(lower=0).round(1)
```

```python
today_ref = pd.Timestamp("2023-12-01")

df["CurrentTenureMonths"] = (
    (today_ref.year - df["HireDate"].dt.year) * 12 +
    (today_ref.month - df["HireDate"].dt.month)
)

def tenure_risk(t):
    if t < 6:
        return 3
    elif t < 24:
        return 2
    elif t < 60:
        return 1
    else:
        return 0

df["TenureRisk"] = df["CurrentTenureMonths"].apply(tenure_risk)
```

```python
def overtime_risk(x):
    if x >= 30:
        return 3
    elif x >= 20:
        return 2
    elif x >= 10:
        return 1
    else:
        return 0

df["OvertimeRisk"] = df["BaseOvertimeHours"].apply(overtime_risk)
```

```python
def performance_risk(p):
    if p == 2:
        return 2
    elif p == 3:
        return 1
    else:
        return 0

df["PerformanceRisk"] = df["BasePerformanceRating"].apply(performance_risk)
```

```python
dept_attrition_risk = {
    "Sales": 2,
    "HR": 0,
    "Technology": 1,
    "Finance": 0,
    "Operations": 2,
    "Support": 2
}

df["DepartmentRisk"] = df["Department"].map(dept_attrition_risk)
```

```python
df["AttritionRiskScore"] = (
    df["TenureRisk"] +
    df["OvertimeRisk"] +
    df["PerformanceRisk"] +
    df["DepartmentRisk"]
)
```

```python
df["AttritionRiskScore"].describe()
```

```python
def exit_probability(score):
    if score >= 8:
        return 0.70
    elif score >= 6:
        return 0.50
    elif score >= 4:
        return 0.30
    elif score >= 2:
        return 0.15
    else:
        return 0.05

df["ExitProb"] = df["AttritionRiskScore"].apply(exit_probability)
```

```python
np.random.seed(42)
df["ExitEmployeeFlag"] = np.random.binomial(1, df["ExitProb"])
```

```python
df["ExitEmployeeFlag"].mean()
```

```python
exit_dates = []

for _, row in df.iterrows():
    if row["ExitEmployeeFlag"] == 1:
        start_date = row["HireDate"] + pd.Timedelta(days=90)
        end_date = pd.Timestamp("2023-12-31")
        
        if start_date >= end_date:
            exit_dates.append(pd.NaT)
            continue
        
        possible_dates = pd.date_range(start_date, end_date, freq="D")
        
        # Higher risk -> more likely to exit earlier
        weights = np.linspace(1.5, 0.5, len(possible_dates)) if row["AttritionRiskScore"] >= 6 else np.ones(len(possible_dates))
        chosen_date = np.random.choice(possible_dates, p=weights / weights.sum())
        exit_dates.append(chosen_date)
    else:
        exit_dates.append(pd.NaT)

df["ExitDate"] = pd.to_datetime(exit_dates)
```

```python
df["TransferEventFlag"] = np.random.binomial(1, 0.08, size=len(df))
```

```python
transfer_dates = []

for _, row in df.iterrows():
    if row["TransferEventFlag"] == 1:
        start_date = max(row["HireDate"] + pd.Timedelta(days=180), pd.Timestamp("2021-01-01"))
        end_date = pd.Timestamp("2023-12-31")
        
        if start_date < end_date:
            possible_dates = pd.date_range(start_date, end_date, freq="D")
            transfer_dates.append(np.random.choice(possible_dates))
        else:
            transfer_dates.append(pd.NaT)
    else:
        transfer_dates.append(pd.NaT)

df["TransferDate"] = pd.to_datetime(transfer_dates)
```

```python
def next_band(b):
    band_order = ["B2", "B3", "B4", "B5", "B6"]
    if b in band_order and b != "B6":
        return band_order[band_order.index(b) + 1]
    return b

df["PromotedBand"] = np.where(
    df["PromotionEventFlag"] == 1,
    df["Band"].apply(next_band),
    df["Band"]
)
```

```python
def promoted_job_title(current_title, promoted):
    if promoted != 1:
        return current_title
    
    mapping = {
        "Analyst": "Senior Analyst",
        "Senior Analyst": "Engineer",
        "Engineer": "Senior Engineer",
        "Senior Engineer": "Manager",
        "Manager": "Manager"
    }
    
    return mapping.get(current_title, current_title)

df["PromotedJobTitle"] = [
    promoted_job_title(title, flag)
    for title, flag in zip(df["JobTitle"], df["PromotionEventFlag"])
]
```

```python
all_departments = ["Sales", "HR", "Technology", "Finance", "Operations", "Support"]

new_departments = []
for _, row in df.iterrows():
    if row["TransferEventFlag"] == 1:
        choices = [d for d in all_departments if d != row["Department"]]
        new_departments.append(np.random.choice(choices))
    else:
        new_departments.append(row["Department"])

df["TransferredDepartment"] = new_departments
```

```python
all_locations = ["Mumbai", "Bangalore", "Hyderabad", "Pune", "Chennai"]

new_locations = []
for _, row in df.iterrows():
    if row["TransferEventFlag"] == 1:
        choices = [l for l in all_locations if l != row["Location"]]
        new_locations.append(np.random.choice(choices))
    else:
        new_locations.append(row["Location"])

df["TransferredLocation"] = new_locations
```

```python
# -------------------------
# REBUILD FINAL ANALYTICAL LAYER
# -------------------------

# TenureMonths (ensure exists)
df_expanded["TenureMonths"] = (
    (df_expanded["SnapshotMonth"].dt.year - df_expanded["HireDate"].dt.year) * 12 +
    (df_expanded["SnapshotMonth"].dt.month - df_expanded["HireDate"].dt.month)
)

# ActiveFlag
df_expanded["ActiveFlag"] = np.where(
    df_expanded["ExitDate"].isna() |
    (df_expanded["SnapshotMonth"] < df_expanded["ExitMonth"]),
    1,
    0
)

# Exit flag
df_expanded["ExitedThisMonthFlag"] = np.where(
    (~df_expanded["ExitMonth"].isna()) &
    (df_expanded["SnapshotMonth"] == df_expanded["ExitMonth"]),
    1,
    0
)

# Join flag
df_expanded["NewJoinerFlag"] = np.where(
    df_expanded["SnapshotMonth"] == df_expanded["HireMonth"],
    1,
    0
)

# Overtime (monthly variation)
np.random.seed(42)
df_expanded["OvertimeHours"] = (
    df_expanded["BaseOvertimeHours"] + np.random.normal(0, 3, size=len(df_expanded))
).clip(lower=0).round(1)

# Performance variation
df_expanded["PerformanceRating"] = (
    df_expanded["BasePerformanceRating"] +
    np.random.choice([-1, 0, 1], size=len(df_expanded), p=[0.1, 0.8, 0.1])
).clip(1, 5)

# LeaveDays
df_expanded["LeaveDays"] = np.random.poisson(1.5, size=len(df_expanded))

# Absenteeism
df_expanded["AbsenteeismRate"] = (df_expanded["LeaveDays"] / 22 * 100).round(1)

# Salary progression
df_expanded["YearsSinceHire"] = np.floor(df_expanded["TenureMonths"] / 12).astype(int)

df_expanded["AnnualIncrementPercent"] = np.random.choice(
    [0.00, 0.05, 0.08, 0.10],
    size=len(df_expanded),
    p=[0.10, 0.35, 0.35, 0.20]
)

df_expanded["SalaryAfterAnnualIncrement"] = (
    df_expanded["BaseSalary"] *
    ((1 + df_expanded["AnnualIncrementPercent"]) ** df_expanded["YearsSinceHire"])
)

df_expanded["PromotionJumpFactor"] = np.where(
    (~df_expanded["PromotionMonth"].isna()) &
    (df_expanded["SnapshotMonth"] >= df_expanded["PromotionMonth"]),
    1.15,
    1.00
)

df_expanded["MonthlyBaseSalary"] = (
    df_expanded["SalaryAfterAnnualIncrement"] *
    df_expanded["PromotionJumpFactor"]
).round(0)

# Increment %
df_expanded["IncrementPercent"] = np.where(
    df_expanded["PromotionFlag"] == 1,
    0.15,
    np.where(
        df_expanded["SnapshotMonth"].dt.month == df_expanded["HireDate"].dt.month,
        df_expanded["AnnualIncrementPercent"],
        0.00
    )
)

# Time since promotion
df_expanded["TimeSinceLastPromotionMonths"] = np.where(
    (~df_expanded["PromotionMonth"].isna()) &
    (df_expanded["SnapshotMonth"] >= df_expanded["PromotionMonth"]),
    (
        (df_expanded["SnapshotMonth"].dt.year - df_expanded["PromotionMonth"].dt.year) * 12 +
        (df_expanded["SnapshotMonth"].dt.month - df_expanded["PromotionMonth"].dt.month)
    ),
    df_expanded["TenureMonths"]
)

# Role reset logic
df_expanded["RoleResetMonth"] = df_expanded[["PromotionMonth", "TransferMonth"]].max(axis=1)

df_expanded["TimeInCurrentRoleMonths"] = np.where(
    df_expanded["RoleResetMonth"].notna(),
    (
        (df_expanded["SnapshotMonth"].dt.year - df_expanded["RoleResetMonth"].dt.year) * 12 +
        (df_expanded["SnapshotMonth"].dt.month - df_expanded["RoleResetMonth"].dt.month)
    ),
    df_expanded["TenureMonths"]
)
```
